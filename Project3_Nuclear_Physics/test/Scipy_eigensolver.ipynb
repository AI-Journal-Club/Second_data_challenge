```python
import sys
sys.path.insert(0, '../include')

import numpy as np
from plot import plot_wavefunction
from scipy_eigensolver import solve_1channel
```

Define grid

```python
r_min = 1e-4
r_max = 30.0
r = np.linspace(r_min, r_max, 1000)
```

Square-well potential

```python
V0 = 59.66
R0 = 1.5

def V(r):
    return np.where(r <= R0, -V0, 0.0)

E, u, R = solve_1channel(V, r)
plot_wavefunction(r, u, R, E)
```

Gaussian potential

```python
V0 = 70.92
R0 = 1.5

def V(r):
    return -V0 * np.exp(-r**2 / R0**2)

E, u, R = solve_1channel(V, r)
plot_wavefunction(r, u, R, E)
```

Yukawa potential

```python
g = 8.22
a = 1.5 
mu = 1.0/a

def V(r):
    return - g**2 * np.exp(-mu * r) /  np.maximum(r, 1e-2)

E, u, R = solve_1channel(V, r)
plot_wavefunction(r, u, R, E)
```
